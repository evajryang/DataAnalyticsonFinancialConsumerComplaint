# Part I Data Description

#### Read data set and skim inside

<p> Consumer of complaints is a data set collected and posted by CFPB(Consumer Financial Protection Bureau).CFPB believes that data from complaints will assist them to understand the marketplace and to protect consumers. In this data set, each row is a record of complaint from consumer, whereas the columns report:
<ul>
<li><b>date received</b>: The date when CFPB has received the complaint.
<li><b>Product</b>: The type of product of which the consumer's complaint is associated.
<li><b>Sub-product</b>: The type of sub-product of which the consumer's complaint is associated 
<li><b>Issue</b>:  The type of issue of which the consumer's complaint is associated.
<li><b>Sub-issue</b>:  The type of sub-issue of which the consumer's complaint is associated.
<li><b>Consumer complaint narrative</b>: Description of consumers' submitted complaint.
<li><b>Company public response</b>: The optional, public-facing response from company.
<li><b>Company</b>: The company about which complaint is about.
<li><b>State</b>: The state of mailing address of consumer.
<li><b>ZIP code</b>: The zip code of mailing address of consumer.
<li><b>Tags</b>: The tags sorting complaints submitted by or on behalf of consumers.
<li><b>Consumer consent provided?</b>: The identifier showing whether the consumer chose to publish complaint narrative text.
<li><b>Submitted via</b>: The way of complaint to be submitted to CFPB.
<li><b>Date sent to company</b>: The date when CFPB sent complaint to company.
<li><b>Company response to consumer</b>: The response from company to consumer.
<li><b>Timely response?</b>: Whether the company responded timely.
<li><b>Consumer disputed?</b>: Whether the consumer disputed the response from the company.
<li><b>Complaint ID</b>: The unique identification number of a complaint.
</ul></p>
<p> For further info, please visit https://cfpb.github.io/api/ccdb/fields.html
<ul>

# Part II Data Cleaning

### Basic cleaning

#### Change columns'names and place Complaint ID at first column

```python
df.columns = ['date_received',
                       'products',
                       'sub_product',
                       'issue',
                       'sub_issue',
                       'consumer_comp_nar',
                       'company_pb_rsps',
                       'company',
                       'state',
                       'zipcode',
                       'tags',
                       'consumer_cons_prov',
                       'submitted_via',
                       'date_sent_to_company',
                       'company_rsps_to_c',
                      'if_timely_rsps',
                      'consumer_disp',
                      'complaint_id']
```

```python
df = df.sort_values(by=['complaint_id'])
df = df.reset_index(drop=True)
```

```python
list_of_columns = df.columns.tolist()
list_of_columns.insert(0, list_of_columns.pop(-1))
df = df.reindex(columns=list_of_columns)
```

```python
df.head(1)
```

#### Check NaN

```python
df.isna().any().any()
```

```python
df.isna().sum()
```

```python
df.head(1)
```

#### Dealing with missing data

```python
df.sub_product.fillna('Unknown',inplace=True)
df.sub_issue.fillna('Unknown',inplace=True)
df.state.fillna('Unknown',inplace=True)
df.zipcode.fillna('Unknown',inplace=True)
```

```python
df.consumer_comp_nar.fillna('None',inplace=True)
df.tags.fillna('None',inplace=True)
```

```python
df.company_pb_rsps.fillna('Company chose not to anwser',inplace=True)
df.consumer_cons_prov.fillna('Other',inplace=True)
df.consumer_disp.fillna('No',inplace=True)
```

```python
df.isna().any().any()
```

```python
df.head(1)
```

#### Replace *Yes* and *No* with 1 and 0

```python
df.replace(to_replace='Yes', value=1.0, inplace=True)
df.replace(to_replace='No', value=0.0, inplace=True)
```

```python
df.head(1)
```

#### Convert date columns from type string to timestamp

```python
df['date_received'] =  pd.to_datetime(df.date_received)
df['date_sent_to_company'] = pd.to_datetime(df.date_sent_to_company)
```

```python
df.head(1)
```

### Advanced Cleaning

#### 1.Add a new column showing days proceeding consumer's complaint from Bureau to company

```python
df['process_send'] = (df.date_sent_to_company - df.date_received).dt.days
```

#### 2.Add a new column to show if complaint is delivered in one day, one week, one month, etc.

```python
df.loc[df['process_send'] <= 7, 'process_period'] = 'in a week'
```

```python
df.loc[(df['process_send']>7)&(df['process_send']<=30),'process_period' ]='in a month'
```

```python
df.loc[(df['process_send']>30)&(df['process_send']<=120),'process_period' ]='in a quarter'
```

```python
df.loc[(df['process_send']>120)&(df['process_send']<=365),'process_period' ]='in a year'
```

```python
df.loc[(df['process_send']>365),'process_period' ]='more than a year'
```

```python
df.head(1)
```

#### 3.Add a new column to show company's geographical distribution level.

```python
company_number = df.groupby('company')['state'].nunique()
```

```python
company_number = pd.DataFrame(company_number)
company_number.reset_index(inplace=True)
```

<p> <b>According to how many different states/territory where the company shows up:</b>
<ul>
<li><b>4</b>: Super-National
<li><b>3</b>: National
<li><b>2</b>: Expanding
<li><b>1</b>: Multi-state
<li><b>0</b>: Single_state
</ul></p>

```python
df = df.merge(company_number, left_on='company',right_on='company',how='left')
```

```python
df.loc[df['state_y'] >50, 'company_map'] = 4
```

```python
df.loc[df['state_y'] ==50, 'company_map'] = 3
```

```python
df.loc[(df['state_y'] <50)&(df['state_y'] >= 25), 'company_map'] = 2
```

```python
df.loc[(df['state_y'] <25)&(df['state_y'] > 1), 'company_map'] = 1
```

```python
df.loc[df['state_y'] ==1, 'company_map'] = 0
```

```python
df.rename(columns = {'state_x':'state'}, inplace = True) 
df.rename(columns = {'state_y':'num_different_state'}, inplace = True) 
```

```python
df.head(1)
```

####  4.Last but not least, drop columns that will not be needed.

```python
df.drop(columns=['zipcode'],inplace=True)
df.drop(columns=['consumer_cons_prov'],inplace=True)
```

Use complaints with narrative in the record.

```python
cf = df[df.consumer_comp_nar != 'None']
```

```python
import string 
import re
cf['split_nar'] = cf.consumer_comp_nar.apply(lambda s: re.sub(r'[^\w\s]','',s).lower().split())
cf['split_nar'] = cf['split_nar'].apply(lambda x: [k for k in x if 'xxxx' not in k])
```

Let's add another to see the length of each list.

```python
cf['len'] = cf['split_nar'].apply(lambda x:len(x))
```

 Use Anaconda to install: *-pip install pyspellchecker* to check strings that seem to have typographical error. 

```python
from spellchecker import SpellChecker
spell = SpellChecker()
cf['misspelled_n'] = cf.split_nar.apply(lambda x: list(spell.unknown(x)))
```

Create a column to show percentage of words that might be typos in each narrative.

```python
cf['misspelled_ratio'] = (cf.misspelled_n.apply(lambda l:len(l)) / cf.split_nar.apply(lambda l:len(l)))*100
```

```python
cf.misspelled_ratio.fillna(0,inplace=True)
```

```python
cf.isna().any().any()
```

We choose to use logistic regression to create decision tree graph.  
First, we choose the variables needed to run classification.

```python
cf1 = cf[['company_rsps_to_c','if_timely_rsps','consumer_disp','process_send','misspelled_ratio','products','len','company_map']]
```

Second, non-numeric variables will be converted to dummy variables.

```python
cf2=pd.get_dummies(cf1,columns=['company_rsps_to_c','products'],dummy_na=True)
```

We might need a deeper depth to see a clear path:

```python
from sklearn.linear_model import LogisticRegression
import sklearn as sk
import sklearn.tree as tree
from IPython.display import Image  
import pydotplus
```

```python
import sklearn.tree
dt = sklearn.tree.DecisionTreeRegressor(max_depth =6)
```

```python
Y = cf2.consumer_disp
X = cf2.drop('consumer_disp', axis=1)
```

```python
dt.fit(X,Y)
```

```python
import sklearn.tree as tree
from IPython.display import Image  
import pydotplus

dt_feature_names = list(X.columns)
dt_target_names = np.array(Y.unique(),dtype=np.string_) 
tree.export_graphviz(dt, out_file='tree.dot', 
    feature_names=dt_feature_names, class_names=dt_target_names,
    filled=True)  
graph = pydotplus.graph_from_dot_file('tree.dot')
Image(graph.create_png())
```

<b>What we have from prediction:  </b>  
While sending a complaint about bank account/service, a consumer whose complaint has greater than 2.05% typos and whose response from a non-national company is ambiguous will very likely dispute.

#### Validations:
<ul>1. The most disputed response from company to consumer is *Closed*.</ul>

```python
gr1 =cf.groupby(['products','company_rsps_to_c'])['consumer_disp'].mean()
sns.heatmap(gr1.unstack(),annot=True,cmap='Blues')
plt.title("Company response VS dispute rate for different Products ", size=18)
```

<ul>2. Company of smaller size is more likely to be disputed.</ul>

```python
cf['small_company']=cf.company_map.apply(lambda x: 1.0 if x<=1.5 else 0.0)
```

```python
gr2 =cf.groupby(['products','small_company'])['consumer_disp'].mean()
sns.heatmap(gr2.unstack(),annot=True,cmap='Greens')
plt.title("if company small VS products VS dispute rate ", size=18)
```

<ul>4. We wonder what makes bank account/service product more likely to be disputed: </ul>

```python
cf_issue = cf.groupby('products')['issue'].nunique().sort_values().reset_index()
cf_sub_issue = cf.groupby('products')['sub_issue'].nunique().sort_values().reset_index()
cf_product = cf.groupby('products')['consumer_disp'].mean().sort_values().reset_index()


sns.catplot(y='products',x='issue',data=cf_issue,kind='bar',palette='Greens',aspect=2)
plt.title("Issue selections for different Products ", size=18)
sns.catplot(y='products',x='sub_issue',data=cf_sub_issue,kind='bar',palette='Blues',aspect=2)
plt.title("Sub-issue selections for different Products ", size=18)
sns.catplot(y='products',x='consumer_disp',data=cf_product,kind='bar',palette='Reds')
plt.title("Average dispute rate for different Products ", size=18)
```

It seems products with less selection in issue/sub-issue are more likely to be disputed...  
Let's make regression plots with the corresponding dispute rate of each product.

```python
cf_reg = cf_issue.merge(cf_sub_issue,on='products').merge(cf_product,on='products')
```

```python
sns.regplot(x='issue',y='consumer_disp',data=cf_reg)
plt.title("Regression: Issue Number VS Dispute rate ", size=18)
```

```python
sns.regplot(x='sub_issue',y='consumer_disp',data=cf_reg)
plt.title("Regression: Sub-issue Number VS Dispute rate ", size=18)
```

#### From above, we can see that products with more issues/sub-issues will be less likely to be disputed.  
  
Since consumers select their issue/sub-issue from a pre-list, they will have to explain what is subjective about which their problem is when the pre-list is lacking enough selections.

### (A)Finding 

<b>Number of issue & sub-issue selections of products are negatively related to dispute rate.</b>  

### (A)Managerial insights

In order to decrease disputes, we suggest:  
<b>To CFPB: Shorten period of sending complaints and import auto-correction module in complaint delivery system. 
                       Make more issue/sub-issue subjects for most disputed products, and thus to avoid the misunderstanding between consumer and company and to save time for both parties.</b>

```python
df_t = pd.read_csv("tuition_cost.csv")
```

#### We define private_class_rate as follow :

<p> 
<ul>
<li>high (private school in state:over 50%)
<li>medium (private school in state:between 30% and 50%)
<li>low (private school in state:below 30%)
</ul></p>

```python
df_stu['private_rate_class'] = df_stu.private_rate.apply(lambda x: 'low' if x <= 0.3 \
                                            else 'medium' if 0.3 < x <= 0.5 \
                                            else 'high')  
```

```python
df_stu.head(1)
```

```python
sns.countplot(x='private_rate_class', data=df_stu,order = df_stu['private_rate_class'].value_counts().index,palette='Spectral')
plt.title("Number of complaints VS Private rate class", size=18)
```

```python
graph3 = df_stu.groupby(['issue','private_rate_class'])\
    ['consumer_disp'].mean()
sns.heatmap(graph3.unstack(),annot=True,cmap='Reds')
plt.title("Private rate VS Issues VS Dispute rate ", size=18)
```

#### From the heat map above, we can know that, among all the issues in the student loan, there is a higher chance to be disputed in the state of low private school rate. 

Let us clustering on the low private rate class.

First, we convert the private rate to  0,1,2.

```python
df_stu.replace(to_replace='low', value=0, inplace=True)
df_stu.replace(to_replace='medium', value=1, inplace=True)
df_stu.replace(to_replace='high', value=2, inplace=True)
```

```python
df_stu_clu = df_stu[df_stu.private_rate_class == 0]
df_stu_clu.head(3)
```

## (C)Managerial insights

#### Borrowers with non-federal student loans have defaulted on federal loans, and they are often in poor financial condition. Therefore, they are likely to delay repayment, leading to conflicts with banks and increasing disputed rates.  In the future, banks need to strictly audit the users who need non federal loans, which can effectively reduce the complaint rate and even reduce the risk of default.
